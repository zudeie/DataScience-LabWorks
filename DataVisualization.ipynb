# Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading dataset

```python

# Load the dataset
df = pd.read_csv("Health.csv")
```

```python
sns.set_theme(style="darkgrid")
```

## Data Visualization

```python
# Figure size setup
fig, axes = plt.subplots(3, 2, figsize=(15, 18))
fig.suptitle("Data Visualizations", fontsize=16)
# Line Chart: Trend of BMI across the dataset index
axes[0, 0].plot(df.index[:100], df["BMI"][:100], marker="o", linestyle="-", color="b")
axes[0, 0].set_title("Line Chart: BMI Trend")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("BMI")

# Bar Plot: Average BMI by sex
sns.barplot(x=["Male", "Female"], y=df.groupby("sex")["BMI"].mean(), ax=axes[0, 1])
axes[0, 1].set_title("Bar Plot: Average BMI by Sex")
axes[0, 1].set_ylabel("Average BMI")

# Pie Chart: Distribution of sex in dataset
sex_counts = df["sex"].value_counts()
axes[1, 0].pie(sex_counts, labels=["Male", "Female"], autopct="%1.1f%%", colors=["blue", "pink"])
axes[1, 0].set_title("Pie Chart: Gender Distribution")

# Box Plot: Distribution of Systolic Blood Pressure (SysBP)
sns.boxplot(y=df["SysBP"], ax=axes[1, 1])
axes[1, 1].set_title("Box Plot: Systolic Blood Pressure Distribution")

# Histogram: Distribution of BMI
sns.histplot(df["BMI"], bins=30, kde=True, ax=axes[2, 0])
axes[2, 0].set_title("Histogram: BMI Distribution")

# Scatter Plot: Relationship between Weight and Height
sns.scatterplot(x=df["weightKg"], y=df["heightCm"], alpha=0.5, ax=axes[2, 1])
axes[2, 1].set_title("Scatter Plot: Weight vs Height")

# Adjust layout and show plots
plt.tight_layout(rect=[0, 0.03, 1, 0.97])
plt.show()
```

# Based on the visualizations:

* Line Chart: BMI Trend  The BMI values fluctuate significantly, showing a wide range across the dataset. There is no clear upward or downward trend, indicating variability in BMI.

* Bar Plot: Average BMI by Sex  The average BMI for males and females appears nearly equal, suggesting that BMI distribution does not differ significantly between genders.

* Pie Chart: Gender Distribution  The dataset has a balanced gender distribution, with approximately 50.1% males and 49.9% females.

* Box Plot: Systolic Blood Pressure Distribution  The median systolic blood pressure is around 130, with an interquartile range spanning approximately 120 to 140. There might be a few outliers below 110.

* Histogram: BMI Distribution  The BMI distribution is slightly right-skewed, with most individuals having a BMI between 20 and 35. There is a peak around 30, which might indicate a high proportion of overweight individuals.

* Scatter Plot: Weight vs Height  The points form a triangular shape, suggesting a strong correlation between height and weight. Taller individuals generally have higher weight values, as expected.
